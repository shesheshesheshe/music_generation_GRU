```python
import vs_new as vs
import utils
import pandas as pd
import numpy as np
import pickle
```

```python
# which dataset are you going to feed into the MUSIC MODEL?
# artist_name_music_model_list options: 'bach', 'cond', 'a_minor', 'scarl', 'auto', 'c_minor'
artist_name_music_model_list = ['cond']

# which prompt song are you going to feed into the MUSIC MODEL?
# prompt_name_list options: any existed .pkl generated by the BAR-2 MODEL can be chosen
prompt_name_list = ['data_generation/07041634_bach_GRU3_L2_BatchNorm_B2G_None_ckp_rg_g_p.pkl']

# what are the parameters for the MUSIC MODEL?
music_model_details = [3, # num of GRU layer?
                       False, # time distribution layer?
                       True, # batch normalization layer?
                       True, # L2 regularization in GRU layers?
                       20, # input_note_num
                       80, # seq_len
                       'nadam' # optimizer
                       ] 
```

```python
# Generate the music by the MUSIC MODEL
i = 0
artist_name_music_model = artist_name_music_model_list[i]
prompt_name = prompt_name_list[i]

# load prompt song
with open(prompt_name, 'rb') as file:
    data = pickle.load(file)

# Start training the MUSIC Model
music_model = vs.music_generation(artist_name=artist_name_music_model, epochs=700, model_details=music_model_details)
music_model.model_learning_stage()

# Start generating music
music_model.generate_data(sequence_num=len(data[3]), prompt=data[2], ans=data[3], ckp_name=None, prompt_type='song')
```
